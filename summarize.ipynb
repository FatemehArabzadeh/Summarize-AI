```python
!pip install google-generativeai

from kaggle_secrets import UserSecretsClient

user_secrets = UserSecretsClient()
api_key = user_secrets.get_secret("GEMINI_API_KEY")

if api_key:
    print("✅ API Key is successfully retrieved!")
else:
    print("❌ API Key still not found. Please check your Kaggle Secrets settings.")

import google.generativeai as genai
from kaggle_secrets import UserSecretsClient


user_secrets = UserSecretsClient()
api_key = user_secrets.get_secret("GEMINI_API_KEY")


genai.configure(api_key=api_key)
model = genai.GenerativeModel("gemini-pro")


def summarize_text_with_prompt_engineering(text):
    prompt = f"لطفاً این متن را به‌طور خلاصه و با حفظ جزئیات اصلی خلاصه کن.:\n\n{text}"
    
    response = model.generate_content(prompt)
    
    return response.text if response else "خلاصه‌ای پیدا نشد."

input_text = """هوش مصنوعی (AI) شاخه‌ای از علوم کامپیوتر است که بر ایجاد سیستم‌هایی تمرکز دارد که می‌توانند مانند انسان فکر کنند، یاد بگیرند و تصمیم‌گیری کنند. AI شامل یادگیری ماشینی، پردازش زبان طبیعی، بینایی کامپیوتری و بسیاری از زیرشاخه‌های دیگر است."""

summary = summarize_text_with_prompt_engineering(input_text)

print("🔹 متن اصلی:")
print(input_text)
print("\n🔹 خلاصه متن:")
print(summary)

!pip install gradio

import gradio as gr

def summarize_ui(text):
    return summarize_text(text)

iface = gr.Interface(
    fn=summarize_ui,
    inputs=gr.Textbox(lines=5, placeholder="Enter text here..."),
    outputs="text",
    title="Text Summarizer",
    description="Enter a text and get a concise summary using Google Gemini API."
)

iface.launch(share=True)
```
